```python
pianofile = 'r1.wav'
piano2file = 'r2.wav'

pianoSig , sr2 = librosa.load(pianofile,mono=True,sr=None)
piano2Sig , sr3 = librosa.load(piano2file,mono=True,sr=None)

timepiano=np.arange(0,pianoSig.shape[0])/sr2
timepiano2=np.arange(0,piano2Sig.shape[0])/sr3

pianoSpec = librosa.amplitude_to_db(np.abs(librosa.stft(pianoSig)), ref=np.max) 
librosa.display.specshow(pianoSpec, y_axis='linear')
plt.colorbar(format='%+2.0f dB')
plt.title('R1: Linear-frequency power spectrogram')
```
