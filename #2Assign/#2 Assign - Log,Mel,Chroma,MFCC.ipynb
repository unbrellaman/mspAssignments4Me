```python
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt
```

```python
voxfile = 'avm.wav'
sig , sr = librosa.load(voxfile,mono=True,sr=None)
```

```python
voxfile = 'avm.wav'
voxSig , sr2 = librosa.load(voxfile,mono=True,sr=None)
timevox=np.arange(0,voxSig.shape[0])/sr
n_fft=2048, 
hop_length=32
win_length=1024
voxSpec = librosa.amplitude_to_db(np.abs(librosa.stft(voxSig)), ref=np.max) 
librosa.display.specshow(voxSpec, y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Log-frequency power spectrogram')
```

```python
S = librosa.feature.melspectrogram(y=sig, sr=sr, n_fft=2048, n_mels=128)
log_S = librosa.power_to_db(S, ref=np.max)
librosa.display.specshow(log_S, sr=sr, x_axis='time', y_axis='mel')
plt.title('Mel-power spectrogram')
plt.colorbar(format='%+2.0f dB')
plt.tight_layout()
```

```python
mfcc = librosa.feature.mfcc(y=sig, sr=sr, hop_length=32, n_mfcc=13)
librosa.display.specshow(mfcc[1:12,:],x_axis='time')
plt.ylabel('MFCCs')
```

```python
# plot a stft spectrogram
regSpec = librosa.amplitude_to_db(np.abs(librosa.stft(sig)), ref=np.max)

plt.figure(figsize=(12,4))

librosa.display.specshow(regSpec, y_axis='log')

plt.title('STFT spectrogram')
```
